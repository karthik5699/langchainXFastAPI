```python
from datetime import datetime
time = []
value = []
data = []

for period in q:
    t = period['date']
    t_obj = datetime.strptime(t, "%a %b %d %Y %H:%M:%S")
    timestamp = int(t_obj.timestamp())
    time.append(timestamp*1000)
    value.append(period['value'])
    data.append([timestamp*1000, period['value']])


print(time)
print(value)
print(data)
```

```python
from datetime import datetime
copy = q
for e in copy:
    date_obj = datetime.strptime(e['date'], "%a %b %d %Y %H:%M:%S")
    formatted_date = date_obj.strftime("%Y-%m-%d")
    e['time'] = formatted_date
    del e['date']
print(copy)
    
```
